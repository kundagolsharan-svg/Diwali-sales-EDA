total amount spent by which gender based on marital Status.

```python
# total amount spent by which gender based on marital Status.
sales_state = df.groupby(['Gender','Marital_Status'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sns.barplot(x = 'Gender', y = 'Amount',hue='Marital_Status', data = sales_state)
```

```python
# total amount spent by which gender based on marital Status.
sales_state = df.groupby(['Gender','Occupation'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sns.barplot(x = 'Gender', y = 'Amount',hue='Occupation',data = sales_state)
```

```python
# total amount spent by which Occupation
sales_state = df.groupby(['Occupation'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sns.barplot(x = 'Occupation', y = 'Amount', data = sales_state)
```

```python
# total number of order based on product Id
sales_state = df.groupby(['Product_ID'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False)
sns.barplot(x = 'Product_ID',y ='Orders',hue = 'Orders', data = sales_state)
```

```python
# create a subplot based on total number of orders and product Id
import seaborn as sns
import matplotlib.pyplot as plt
sales_state = df.groupby(['Product_ID'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False)
sns.subplot(x = 'Product_ID',y ='Orders',hue = 'Orders', data = sales_state)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Group by Product_ID and sum orders, then sort and get top 10
sales_product = df.groupby(['Product_ID'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)

# Create a figure and a single subplot
fig, ax = plt.subplots(figsize=(12, 6))

# Create the bar plot on the specified subplot
sns.barplot(x='Product_ID', y='Orders', data=sales_product, ax=ax, palette='viridis')

# Add labels and title for clarity
ax.set_xlabel('Product ID')
ax.set_ylabel('Total Orders')
ax.set_title('Top 10 Products by Total Orders')

# Rotate x-axis labels if they are too long
ax.tick_params(axis='x', rotation=45)

# Display the plot
plt.tight_layout()
plt.show()
```
